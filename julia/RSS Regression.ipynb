## Regression Example

```julia
using DataFrames
using GLM
using Gadfly
using Distributions
```

```julia
df = readtable("data/regression.csv")
```

```julia
plot(df, x="Age", y="OI", color="Sex")
```

Data contains negative Age. Select only valid rows. 

```julia
df = df[df[:Age] .> 0, :];
```

```julia
plot(df, x="Age", y="OI", color="Sex")
```

Convert the _Sex_ column to a `PooledDataArray` to signal to the model that this is a categorical variable. 

```julia
pool!(df, [:Sex])
```

```julia
df[:log_OI]=log(df[:OI]);
df
```

```julia
m = lm(log_OI ~ Age + Sex , df)
```

Copy residuals to DataFrame for ease of plotting. Not absolutely necessary. 

```julia
df[:residuals] = residuals(m);
df[:fitted] = predict(m);
df
```

```julia
#Implement QQ plots using Gadfly
import Gadfly.ElementOrFunction
Gadfly.plot(qq::QQPair, elements::ElementOrFunction...; mappings...) = Gadfly.plot(x=qq.qx, y=qq.qy, Geom.point, Theme(highlight_width=0px), elements...; mappings...)
qqplot(x, y, elements::ElementOrFunction...; mappings...) = Gadfly.plot(qqbuild(x, y), elements...; mappings...)
qqnorm(x, elements::ElementOrFunction...; mappings...) = qqplot(Normal(), x, Guide.xlabel("Theoretical Normal quantiles"), Guide.ylabel("Observed quantiles"), elements...; mappings... )
```

```julia
qqnorm(residuals(m), color=df[:Sex][sortperm(residuals(m))])
```
